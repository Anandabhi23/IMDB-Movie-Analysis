## Classify Movies Based on Rating[Excellent , Good & Averege] 

```python
data.columns
```

```python
def rating(rating):
    if rating >= 7.0:
        return "Excellent"
    elif rating>= 6.0:
        return "Good"
    else :
        return "Average"
```

```python
data['rating_cat'] = data['Rating'].apply(rating)
```
